```python
G = nx.Graph()

G = nx.DiGraph()
G.add_node(1)
G.add_node(1)
G.add_nodes_from([1,1,1,4])

G.add_edge(1, 3)
G.add_edges_from([(3, 4), (3, 6), (6, 7)])
nx.draw(G)
plt.show()
```
